# Data Cleaning and Data Prepocessing

##  Check null values in dataset

```python
missing_values = df.isnull().sum()
missing_values
```

```python
df['Current IPLTeam'] = df['Current IPL Team'].fillna('Unknown')
```

## Group players by country and count the number of players from each country

```python
players_by_country = df['Country'].value_counts()
print("Number of Players by Country:")
display(players_by_country)
```

## How is the distribution of player's ages in the IPL Auction 2025?

```python
plt.figure(figsize=(9, 6))
sns.histplot(df['Age'], bins=15, kde=True) #(Kernel Density Estimation)To visualize the underlying distribution of continuous variables.
plt.title('Distribution of Player Ages')
plt.xlabel('Age')
plt.ylabel('Number of Players')
plt.show()
```

### we can observe the auction includes middle aged players

## Analyze the average reserve price by specialism (role)

```python
average_price_by_role = df.groupby('Specialism')['Reserve Price Rs Lakh'].mean().sort_values(ascending=False)
print("Average Reserve Price by Specialism:")
display(average_price_by_role)
```

## Which player had the highest auction price (Price Rs Lakh) in the dataset?

```python
highest_price_player = df.loc[df['Reserve Price Rs Lakh'].idxmax()]
highest_price_player
```

## Which country has the most players in the IPL Auction 2025?

```python
country_count = df['Country'].value_counts()
```

```python
plt.figure(figsize=(9, 6))
country_count.plot(kind='bar', color='teal')
plt.title('Number of Players by Country in IPL Auction 2025')
plt.xlabel('Country')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

### Indian Country has most player in IPL Auction 2025

## Is there any correlation between the reserve price and the player's age in the IPL Auction 2025?

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Reserve Price Rs Lakh'], color='orange', alpha=0.7)
plt.title('Reserve Price vs Age in IPL Auction 2025')
plt.xlabel('Age')
plt.ylabel('Reserve Price (Rs Lakh)')
plt.grid(True)
plt.show()
```

### older players are valued higher due to experience or past performances.

## Who are the top 6 players with the highest reserve prices in the IPL Auction 2025?

```python
top_6_players = df.nlargest(6, 'Reserve Price Rs Lakh')[['First Name', 'Surname', 'Reserve Price Rs Lakh']]
```

```python
plt.figure(figsize=(9, 6))
plt.barh(top_6_players['First Name'] + ' ' + top_6_players['Surname'], top_6_players['Reserve Price Rs Lakh'], color='purple')
plt.title('Top 6 Players with Highest Reserve Prices in IPL Auction 2025')
plt.xlabel('Reserve Price (Rs Lakh)')
plt.ylabel('Player')
plt.show()
```

## How many players specialize in batting, bowling, and all-rounder roles in the IPL Auction 2025?

```python
specialism_count = df['Specialism'].value_counts()
```

```python
# Plotting the distribution of players by Specialism
plt.figure(figsize=(4, 4))
plt.pie(specialism_count, labels=specialism_count.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'lightcoral','lightpink'])
plt.title('Distribution of Players by Specialism in IPL Auction 2025')
plt.show()
```

### 23.8 % players specialize in batting,36.2% in bowling,15.4% wicket keeping and 24.6 % all-rounder roles in the IPL Auction 2025.Bowling skill is more in demand

## Which countries have the highest average reserve prices for their players in the IPL Auction 2025?

```python
avg_reserve_by_country = df.groupby('Country')['Reserve Price Rs Lakh'].mean().nlargest(5)
```

```python
plt.figure(figsize=(6, 4))
avg_reserve_by_country.plot(kind='bar', color='darkblue')
plt.title('Top 5 Countries with Highest Average Reserve Price in IPL Auction 2025')
plt.xlabel('Country')
plt.ylabel('Average Reserve Price (Rs Lakh)')
plt.xticks(rotation=45)
plt.show()
```

### England,New Zealand,Australia,Sri Lanka,Afghanistan countries have the highest average reserve prices for their players in the IPL Auction 2025.These countrie's players are considered the most valuable in the Auction

## How many capped and uncapped players are there in the IPL Auction 2025?

```python
capped_uncapped_count = df['Capped or Uncapped'].value_counts()
```

```python
plt.figure(figsize=(6, 6))
capped_uncapped_count.plot(kind='bar', color=['green', 'red'],width=0.4)
plt.title('Number of Capped vs Uncapped Players in IPL Auction 2025')
plt.xlabel('Player Type')
plt.ylabel('Number of Players')
plt.xticks(rotation=0)
plt.show()
```

### about 79 players already experienced in international cricket compared to those making their first appearance. 

## What is the distribution of players by batting side (left-handed vs right-handed)?

```python
batting_side_count = df['Batting Side '].value_counts()
batting_side_count.plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(4,4))
plt.title('Distribution of Players by Batting Side (Left vs Right) in IPL Auction 2025')
plt.ylabel('')
plt.show()
```
